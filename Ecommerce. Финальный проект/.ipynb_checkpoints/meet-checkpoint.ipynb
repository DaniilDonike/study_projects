# Финальный проект. E-commerse. Выявление профилей потребления.

## Введение

### Цель иследования
Создание медотики сегментирования пользователей для увеличения выручки путем улучшение пользовательского опыта с помощью пресонализированных предложений.

### Описание данных

Датасет описывает транзакции интернет-магазина товаров для дома и быта «Пока
все ещё тут». <br>
Колонки в ecommerce_dataset_upd.csv :
- date — дата заказа;
- customer_id — идентификатор покупателя;
- order_id — идентификатор заказа;
- product — наименование товара;
- quantity — количество товара в заказе;
- price — цена товара.


### План работы

1. Подготовка и изучение данных:
    - Подключение библиотек
    - Загрузка и знакомство с данными
    - Проверка дубликатов
    - Проверка пропусков
    - Типы данных
    - Ошибки данных
    - Вывод
2. Иследовательский анализ данных:
    - Средний чек
    - Выручка с покупателя
    - Самые популярные товары
    - Сымые популярыные товары по сезонам
    - Среднее количество товаров в заказе
3. Сегментация пользователей:
    - По среднему чеку
    - По частоте покупок
    - По количеству товаров в заказе
    - Вывод
4. Проверка статистических гипотез:
    - Статистическая значимость различий выручки по частоте покупок
    - Статистическая занчимость различий среднего чека по количесту товаров
5. Рекомендации
6. Вывод
