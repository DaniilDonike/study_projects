## 2. Иследовательский анализ данных

### Выручка

Для начала посмотрим средний чек по годам. Для этого в таблицу добавим столбец с годом. Также добавим столбец с месяцами и стлбец с выручкой.

```python
ecommerce['year'] = ecommerce['date'].dt.year
ecommerce['month'] = ecommerce['date'].dt.month
ecommerce['revenue'] = ecommerce['quantity'] * ecommerce['price']
```

Создадим сводную таблицу с сумой заказов 

```python
ecom_sum_year = ecommerce.pivot_table(index=['year','order_id'], values = 'revenue', aggfunc='sum').reset_index()
ecom_sum_year.head()
```

Строим график

```python
sns.barplot(x=ecom_sum_year['year'], y=ecom_sum_year['revenue'])
plt.show()
```

Хотя данные за 2018 год у нас не полные, в этом году выручка была самой высокой. 2019 год был чуть похуже, а 2020  имеет самый маленькую выручку. Возможно ковид дает о себе знать. Далее посмотрим выручку по месяцам.

```python
ecom_sum_month = ecommerce.pivot_table(index=['month','order_id'], values = 'revenue', aggfunc='sum').reset_index()
```

```python
sns.barplot(x=ecom_sum_month['month'], y=ecom_sum_month['revenue'])
plt.show()
```

Здесь сильно выделяеться июнь. Это может быть связано с началом всеобщих поездок на дачи. Далее посмотрим выручку по сезонам. Для этого напишем функцию которая создаст нам столбец с временами года.

```python
def seas(mon): # Функция которая присваевает время года исходя из номера месяца
    if mon == 1 or mon == 2 or mon == 12:
        return 'winter'
    elif mon == 3 or mon == 4 or mon ==  5:
        return 'spring'
    elif mon == 6 or mon == 7 or mon == 8:
        return 'summer'
    elif mon == 9 or mon == 10 or mon == 11:
        return 'autumn'
```

```python
ecommerce['season'] = ecommerce['month'].apply(seas) # пременяем функцию
```

```python
ecom_sum_season = ecommerce.pivot_table(index=['season','order_id'], values = 'revenue', aggfunc='sum').reset_index()
```

```python
sns.barplot(x=ecom_sum_season['season'], y=ecom_sum_season['revenue'])
plt.show()
```

По среднему чеку лето лидирует, опять же это может быть связано с дачным сезоном. Далее весна, также возможно из-за начала дачного сезона.

## 3. Сегментация пользователей

### Сегментация по среднему чеку

Проведем сегментацию по среднему чеку среди пользователей.

```python
avg_check = ecommerce.pivot_table(index = 'customer_id', values = ['revenue','order_id'], aggfunc={'revenue':'sum','order_id':'nunique'}).reset_index()
```

```python
avg_check['avg'] = avg_check['revenue'] / avg_check['order_id']
```

```python
def avg(check): # Функция которая сегментирует по среднему чеку
    if check < 5000:
        return 'маленький'
    elif check > 5000 and check < 20000:
        return 'средний'
    elif check > 20000:
        return 'большой'
```

```python
avg_check['segment'] = avg_check['avg'].apply(avg) # Пременяем функцию
```

Строим график

```python
plt.subplots(figsize = (7,7))
sns.scatterplot(x = avg_check['customer_id'], y = avg_check['avg'])
plt.xticks([])
plt.show()
```

В основном средний чек у большенства пользователей держится до 10000. 

```python
avg_check_count = avg_check.pivot_table(index = 'segment', values = 'order_id',aggfunc='count').reset_index()
plt.pie(avg_check_count['order_id'], labels = avg_check_count['segment'])
plt.show()
```

Самая большая группа это маленький средний чек. Она довольно слиьно доминирует над другими. Это может быть связано достаточно с большим количеством факторов. Например со асортиментом или ценообразованием.

### Сегментация по количеству покупок

```python
count_purchase = ecommerce.pivot_table(index = 'customer_id', values = 'order_id', aggfunc='nunique').reset_index()
def cnt(pur): # Функция которая сементирует по количеству покупок
    if pur == 1:
        return 'одна покупка'
    elif pur > 1 and pur <= 3:
        return 'малое количество покупок'
    elif pur >= 4 and pur < 10:
        return 'среднее количество покупок'
    elif pur >= 10 and pur <= 30:
        return 'большое количество покупок'
    elif pur >=31:
        return 'постоянный покупатель'
count_purchase['segment'] = count_purchase['order_id'].apply(cnt)
```

```python
plt.subplots(figsize = (7,7))
sns.scatterplot(x = count_purchase['customer_id'], y = count_purchase['order_id'])
plt.xticks([])
plt.show()
```

Люди покупают до 3 раз. Все что выше редкость. Ниже сторим график по сегментам.

```python
purchase_count = count_purchase.pivot_table(index = 'segment', values = 'order_id',aggfunc='count').reset_index()
plt.subplots(figsize = (15,8))
sns.barplot(x=purchase_count['segment'], y=purchase_count['order_id'])
plt.show()
```

Как видно больше всего тех кто купил один раз. То есть люди не возвращаются обратно. Далее люди купившие до 3 раз. Это уже лучше чем один, но все же не то. Если люди покупают что-то для дачи, то это значит что они готовы покупать постояно. Так как в доме всегда что-то нужно. Три оставшиеся группы почти не видны на таком маштабе. Взглянем более детально.

```python
plt.subplots(figsize = (15,8))
sns.barplot(x=purchase_count['segment'], y=purchase_count['order_id'])
plt.ylim(0,10)
plt.show()
```

Как не странно постоянных покупателей больше чем тех кто просто купил много. Возможно с лояльностью покупателей все не так плохо. Мое мнение что это самое важное сегментирование. Именно те клиенты которые покупают регулярно самые ценные.

### Сегментация по количеству товаров.

Также посмотрим сегментацию по количеству товаров

```python
count_item = ecommerce.pivot_table(index = 'customer_id', values = 'quantity',aggfunc = 'sum').reset_index()
```

```python
def cnti(item): # Функция которая сегментирует по количеству товара
    if item < 10:
        return 'очень малое количество товара'
    elif item >= 10 and item <= 50:
        return 'малое количество товара'
    elif item >= 51 and item <= 100:
        return 'среднее количество товара'
    elif item >= 101 and item <= 300:
        return 'большое количество товара'
    elif item >=301:
        return 'очень бльшое количество товара'
count_item['segment'] = count_item['quantity'].apply(cnti)
```

```python
plt.subplots(figsize = (10,10))
sns.scatterplot(x = count_item['customer_id'], y = count_item['quantity'])
plt.xticks([])
plt.show()
```

Люди покупают до 50 товаров.

```python
item_count = count_item.pivot_table(index = 'segment', values = 'quantity',aggfunc='count').reset_index()
plt.subplots(figsize = (20,8))
sns.barplot(x=item_count['segment'], y=item_count['quantity'])
plt.show()
```

Самая большая группа это те кто купил до 50 товаров.

```python
plt.subplots(figsize = (15,8))
sns.barplot(x=item_count['segment'], y=item_count['quantity'])
plt.ylim(0,100)
plt.show()
```

## 4. Рекомендации

В первую очередь считаю что нужно ориентироваться на сегментацию по количеству совершения покупок. Для нас должны быть в приоритете клиенты которые регулярно совершают покупки, и нужно стремится их количество наращивать. Ведь повторные покупки играют важную роль в успехе бизнеса и имеют значительное влияние на его долгосрочную устойчивость и прибыльность.  Вот несколько причин, почему повторные покупки так важны:

- Увеличение выручки: Повторные покупки способствуют увеличению общей выручки магазина или компании. Когда покупатель возвращается и совершает дополнительные покупки, это добавляет дополнительные доходы, не требуя больших затрат на маркетинг и привлечение новых клиентов.

- Снижение затрат на маркетинг: Привлечение новых клиентов обычно требует больших затрат на маркетинг и рекламу. В то время как у вас уже есть база клиентов, повторные покупки позволяют снизить затраты на маркетинг и фокусироваться на удержании и увеличении активности существующих клиентов.<br>

Привлечение покупателей к повторным покупкам является важной частью стратегии удержания клиентов. Делать это можно разными способами в первую очередь это конечно персонализированный маркетинг и программы лояльности.

Далее по важности можно выделить средний чек. Он напрямую влияет на общую выручку. Что нас собственно и интрересовало, но лояльных и постоянных клиентов легче склонить к более активным тратам чем других покупателей. Поэтому я и выбрал количество покупок как основную сегментацию
