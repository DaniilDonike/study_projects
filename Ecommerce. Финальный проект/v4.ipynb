### –í—ã–≤–æ–¥

–ò—Ç–æ–≥–æ –≤ –¥–∞–Ω–Ω—ã–∑ –Ω–µ –±—ã–ª–æ –ø—Ä–æ–ø—É—Å–∫–æ–≤ –∏–ª–∏ –ø–æ–ª–Ω—ã—Ö –¥—É–±–ª–∏–∫–∞—Ç–æ–≤. –ë—ã–ª–æ –Ω–µ–∫–æ—Ç–æ—Ä–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –≤—ã–±—Ä–æ—Å–æ–≤, –∫–æ—Ç–æ—Ä—ã–µ –º—ã –ø–æ—á–∏—Å—Ç–∏–ª–∏. –î–∞–ª–µ–µ –ø–µ—Ä–µ—Ö–æ–¥–∏–º –∫ –∏—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å—Å–∫–æ–º—É –∞–Ω–∞–ª–∏–∑—É.

## 2. –ò—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–π –∞–Ω–∞–ª–∏–∑ –¥–∞–Ω–Ω—ã—Ö

### –í—ã—Ä—É—á–∫–∞

–î–ª—è –Ω–∞—á–∞–ª–∞ –ø–æ—Å–º–æ—Ç—Ä–∏–º –≤—ã—Ä—É—á–∫—É –ø–æ –≥–æ–¥–∞–º. –î–ª—è —ç—Ç–æ–≥–æ –≤ —Ç–∞–±–ª–∏—Ü—É –¥–æ–±–∞–≤–∏–º —Å—Ç–æ–ª–±–µ—Ü —Å –≥–æ–¥–æ–º. –¢–∞–∫–∂–µ –¥–æ–±–∞–≤–∏–º —Å—Ç–æ–ª–±–µ—Ü —Å –º–µ—Å—è—Ü–∞–º–∏ –∏ —Å—Ç–ª–±–µ—Ü —Å –≤—ã—Ä—É—á–∫–æ–π.

```python
ecommerce['year'] = ecommerce['date'].dt.year
ecommerce['month'] = ecommerce['date'].dt.month
ecommerce['revenue'] = ecommerce['quantity'] * ecommerce['price']
```

–°–æ–∑–¥–∞–¥–∏–º —Å–≤–æ–¥–Ω—É—é —Ç–∞–±–ª–∏—Ü—É —Å –≤—ã—Ä—É—á–∫–æ–π

```python
ecom_sum_year = ecommerce.pivot_table(index=['year'], values = 'revenue', aggfunc='sum').reset_index()
ecom_sum_year.head()
```

–°—Ç—Ä–æ–∏–º –≥—Ä–∞—Ñ–∏–∫

```python
sns.barplot(x=ecom_sum_year['year'], y=ecom_sum_year['revenue'])
plt.show()
```

```python
def seas(mon): # –§—É–Ω–∫—Ü–∏—è –∫–æ—Ç–æ—Ä–∞—è –ø—Ä–∏—Å–≤–∞–µ–≤–∞–µ—Ç –≤—Ä–µ–º—è –≥–æ–¥–∞ –∏—Å—Ö–æ–¥—è –∏–∑ –Ω–æ–º–µ—Ä–∞ –º–µ—Å—è—Ü–∞
    if mon == 1 or mon == 2 or mon == 12:
        return 'winter'
    elif mon == 3 or mon == 4 or mon ==  5:
        return 'spring'
    elif mon == 6 or mon == 7 or mon == 8:
        return 'summer'
    elif mon == 9 or mon == 10 or mon == 11:
        return 'autumn'
```

```python
avg_check = ecommerce.pivot_table(index = 'customer_id', values = ['revenue','order_id'], aggfunc={'revenue':'sum','order_id':'nunique'}).reset_index()
```

```python
avg_check['avg'] = avg_check['revenue'] / avg_check['order_id']
```

```python
def avg(check): # –§—É–Ω–∫—Ü–∏—è –∫–æ—Ç–æ—Ä–∞—è —Å–µ–≥–º–µ–Ω—Ç–∏—Ä—É–µ—Ç –ø–æ —Å—Ä–µ–¥–Ω–µ–º—É —á–µ–∫—É
    if check < 5000:
        return '–º–∞–ª–µ–Ω—å–∫–∏–π'
    elif check > 5000 and check < 20000:
        return '—Å—Ä–µ–¥–Ω–∏–π'
    elif check > 20000:
        return '–±–æ–ª—å—à–æ–π'
```

```python
avg_check['segment'] = avg_check['avg'].apply(avg) # –ü—Ä–µ–º–µ–Ω—è–µ–º —Ñ—É–Ω–∫—Ü–∏—é
```

–°—Ç—Ä–æ–∏–º –≥—Ä–∞—Ñ–∏–∫

```python
plt.subplots(figsize = (7,7))
sns.scatterplot(x = avg_check['customer_id'], y = avg_check['avg'])
plt.xticks([])
plt.show()
```

–í –æ—Å–Ω–æ–≤–Ω–æ–º —Å—Ä–µ–¥–Ω–∏–π —á–µ–∫ —É –±–æ–ª—å—à–µ–Ω—Å—Ç–≤–∞ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –¥–µ—Ä–∂–∏—Ç—Å—è –¥–æ 10000. 

```python
avg_check_count = avg_check.pivot_table(index = 'segment', values = 'order_id',aggfunc='count').reset_index()
plt.pie(avg_check_count['order_id'], labels = avg_check_count['segment'])
plt.show()
```

### –°–µ–≥–º–µ–Ω—Ç–∞—Ü–∏—è –ø–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤—É –ø–æ–∫—É–ø–æ–∫

```python
count_purchase = ecommerce.pivot_table(index = 'customer_id', values = 'order_id', aggfunc='nunique').reset_index()
def cnt(pur): # –§—É–Ω–∫—Ü–∏—è –∫–æ—Ç–æ—Ä–∞—è —Å–µ–º–µ–Ω—Ç–∏—Ä—É–µ—Ç –ø–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤—É –ø–æ–∫—É–ø–æ–∫
    if pur == 1:
        return '–æ–¥–Ω–∞ –ø–æ–∫—É–ø–∫–∞'
    elif pur > 1 and pur <= 3:
        return '–º–∞–ª–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ–∫—É–ø–æ–∫'
    elif pur >= 4 and pur < 10:
        return '—Å—Ä–µ–¥–Ω–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ–∫—É–ø–æ–∫'
    elif pur >= 10 and pur <= 30:
        return '–±–æ–ª—å—à–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ–∫—É–ø–æ–∫'
    elif pur >=31:
        return '–ø–æ—Å—Ç–æ—è–Ω–Ω—ã–π –ø–æ–∫—É–ø–∞—Ç–µ–ª—å'
count_purchase['segment'] = count_purchase['order_id'].apply(cnt)
```

```python
def cnti(item): # –§—É–Ω–∫—Ü–∏—è –∫–æ—Ç–æ—Ä–∞—è —Å–µ–≥–º–µ–Ω—Ç–∏—Ä—É–µ—Ç –ø–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤—É —Ç–æ–≤–∞—Ä–∞
    if item < 10:
        return '–æ—á–µ–Ω—å –º–∞–ª–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ç–æ–≤–∞—Ä–∞'
    elif item >= 10 and item <= 50:
        return '–º–∞–ª–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ç–æ–≤–∞—Ä–∞'
    elif item >= 51 and item <= 100:
        return '—Å—Ä–µ–¥–Ω–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ç–æ–≤–∞—Ä–∞'
    elif item >= 101 and item <= 300:
        return '–±–æ–ª—å—à–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ç–æ–≤–∞—Ä–∞'
    elif item >=301:
        return '–æ—á–µ–Ω—å –±–ª—å—à–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ç–æ–≤–∞—Ä–∞'
count_item['segment'] = count_item['quantity'].apply(cnti)
```
