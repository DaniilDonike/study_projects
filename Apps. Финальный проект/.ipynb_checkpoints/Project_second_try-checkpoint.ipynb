# Финальный проект. Декомпозиция

## Описание проекта

Датасет содержит данные о событиях, совершенных в мобильном приложении "Ненужные вещи". В нем пользователи продают свои ненужные вещи, размещая их на доске объявлений.
В датасете содержатся данные пользователей, впервые совершивших действия в
приложении после 7 октября 2019 года.

Колонки в /datasets/mobile_sources.csv :
- userId — идентификатор пользователя,
- source — источник, с которого пользователь установил приложение.
<br>

Колонки в /datasets/mobile_dataset.csv :
- event.time — время совершения,
- user.id — идентификатор пользователя,
- event.name — действие пользователя.
<br>

Виды действий:
- advert_open — открыл карточки объявления,
- photos_show — просмотрел фотографий в объявлении,
- tips_show — увидел рекомендованные объявления,
- tips_click — кликнул по рекомендованному объявлению,
- contacts_show и show_contacts — посмотрел номер телефона,
- contacts_call — позвонил по номеру из объявления,
- map — открыл карту объявлений,
- search_1 — search_7 — разные действия, связанные с поиском по сайту,
- favorites_add — добавил объявление в избранное.

## Задачи проекта

1. Проанализировать связь целевого события — просмотра контактов — и других действий пользователей.
2. Оценить, какие действия чаще совершают те пользователи, которые просматривают контакты.

## Цели

Анализ поведения пользователей приложения "Ненужные вещи" для увеличения конверсии в целевое действие, что в свою очередь увеличит вовлеченость

### Время проведеное в приложении

Для начала посмотрим сколько всего времени провели пользователи в нашем приложении за все время. И в первую очередь узнаем  данными за какой временой промежуток мы распологаем.

```python
df['event_time'].min()
```

```python
df['event_time'].max()
```

Итак с 07.10.2019 по 03.11.2019. 28 дней или 4 недели. Добавим столбец в котором будет только дата, без времени.

```python
df['event_date'] = pd.to_datetime(dataset['event_time']).dt.round("d") # добвавляем столбец с датой
```

Далее приступаем к выделению сессий. Для этог воспользуемся способом который был в презентации, соответсвенно сессии выделяем сессии по тайм-ауту. Выбор временного интервала для определения сессий (в данном случае 20 минут) когда пользователь не проявляет активности в течение этого времени, это может сигнализировать о том, что он завершил текущую сессию. 
- 20 минут кажетутся рациональным выбором так как пользователя может что-то отвлечь от приложения.  Слишком короткий тайм-айт может создать много лишних сессий в данных. 
- Длинее тоже не лучшая идея. Тайм-аут более 20 минут слишком большой для специфики мобильного приложения. Ведь люди редко оставляют свои устройства на такое больщое время без присмотра  
<br>

Вот и выходит что выбор 20 минутного тайм-аута основан балансе и специфики мобильного приложения.

```python
df = df.sort_values(['user_id', 'event_time']) # сортируем
```

```python
get_sessions = (df.groupby('user_id')['event_time'].diff() >= pd.Timedelta('20Min')).cumsum() # Этот код взят из пачки
df['session_id'] = df.groupby(['user_id', get_sessions], sort=False).ngroup() + 1
df
```

Столбец с id сессии добавлен. Теперь создадим таблицу в которой будут id сессии, дата начала и конца сессии, а также ее продолжительность

```python
session_time = df.groupby('session_id')['event_time'].agg(['min','max']) # Строим таблицу с датой начала и конца сессии
session_time['time'] = round((session_time['max'] - session_time['min']).dt.seconds/60,0) # добавляем столбец сколько минут длилась сессия
session_time = session_time.reset_index() 
session_time
```

```python
# код взять из урока по расчету ratention rate в python

def get_profiles(df):

    # сортируем сессии по ID пользователя и дате посещения
    # группируем по ID и находим первые значения session_start и channel
    # столбец с временем первого посещения назовём first_ts
    # от англ. first timestamp — первая временная отметка
    profiles = (
        df.sort_values(by=['user_id', 'event_time'])
        .groupby('user_id')
        .agg({'event_time': 'first', 'source': 'first','event_name' : 'first'})
        .rename(columns={'event_time': 'first_ts'})
        .reset_index()  # возвращаем user_id из индекса
    )

    # определяем дату первого посещения
    # и первый день месяца, в который это посещение произошло
    # эти данные понадобятся для когортного анализа
    profiles['date'] = pd.to_datetime(profiles['first_ts'].dt.strftime('%Y-%m-%d')) 
    #profiles['week'] = profiles['first_ts'].dt.week
    profiles['week'] = profiles['first_ts'].dt.isocalendar().week
    
    return profiles
```

```python
# расчитываем Retention rate
def get_retention(
    profiles, df, observation_date, horizon_days, ignore_horizon=False
):

    # исключаем пользователей, не «доживших» до горизонта анализа
    last_suitable_acquisition_date = observation_date
    if not ignore_horizon:
        last_suitable_acquisition_date = observation_date - timedelta(
            days=horizon_days - 1
        )
    result_raw = profiles.query('date <= @last_suitable_acquisition_date')

    # собираем «сырые» данные для расчёта удержания
    result_raw = result_raw.merge(
        df[['user_id', 'event_time']], on='user_id', how='left'
    )
    result_raw['lifetime'] = (
        result_raw['event_time'] - result_raw['first_ts']
    ).dt.days

    # рассчитываем удержание
    result_grouped = result_raw.pivot_table(
        index=['date'], columns='lifetime', values='user_id', aggfunc='nunique'
    )
    cohort_sizes = (
        result_raw.groupby('date')
        .agg({'user_id': 'nunique'})
        .rename(columns={'user_id': 'cohort_size'})
    )
    result_grouped = cohort_sizes.merge(
        result_grouped, on='date', how='left'
    ).fillna(0)
    result_grouped = result_grouped.div(result_grouped['cohort_size'], axis=0)

    # исключаем все лайфтаймы, превышающие горизонт анализа
    result_grouped = result_grouped[
        ['cohort_size'] + list(range(horizon_days))
    ]

    # восстанавливаем столбец с размерами когорт
    result_grouped['cohort_size'] = cohort_sizes

    # возвращаем таблицу удержания и сырые данные
    # сырые данные пригодятся, если нужно будет отыскать ошибку в расчётах
    return result_raw, result_grouped
```

Вызываем написаную функцию с необходимыми параметрами

Так как сессии мы уже выделели на этапе когда изучали время поведенное в приложении, начинаем с веделения сценариев использования. Попробуем реализовать диаграму сэнкей. Этот код взять с хабра

```python
def add_features(df):
    """Функция генерации новых столбцов для исходной таблицы

    Args:
        df (pd.DataFrame): исходная таблица.
    Returns:
        pd.DataFrame: таблица с новыми признаками.
    """
    
    # сортируем по id и времени
    sorted_df = df.sort_values(by=['session_id', 'event_time']).copy()
    # добавляем шаги событий

    
    # удаляем дубликаты событий внутри каждой сессии
    sorted_df = sorted_df.drop_duplicates(subset=['session_id', 'event_name'], keep='first')
    sorted_df['step'] = sorted_df.groupby('session_id').cumcount() + 1
    # добавляем узлы-источники и целевые узлы
    # узлы-источники - это сами события
    sorted_df['source'] = sorted_df['event_name']
    # добавляем целевые узлы
    sorted_df['target'] = sorted_df.groupby('session_id')['source'].shift(-1)
    
    # возврат таблицы без имени событий
    return sorted_df.drop(['event_name'], axis=1)

# преобразуем таблицу
table = add_features(df)
table.head()
```

```python
def get_source_index(df):
    
    """Функция генерации индексов source

    Args:
        df (pd.DataFrame): исходная таблица с признаками step, source, target.
    Returns:
        dict: словарь с индексами, именами и соответсвиями индексов именам source.
    """
    
    res_dict = {}
    
    count = 0
    # получаем индексы источников
    for no, step in enumerate(df['step'].unique().tolist()):
        # получаем уникальные наименования для шага
        res_dict[no+1] = {}
        res_dict[no+1]['sources'] = df[df['step'] == step]['source'].unique().tolist()
        res_dict[no+1]['sources_index'] = []
        for i in range(len(res_dict[no+1]['sources'])):
            res_dict[no+1]['sources_index'].append(count)
            count += 1
            
    # соединим списки
    for key in res_dict:
        res_dict[key]['sources_dict'] = {}
        for name, no in zip(res_dict[key]['sources'], res_dict[key]['sources_index']):
            res_dict[key]['sources_dict'][name] = no
    return res_dict
  

# создаем словарь
source_indexes = get_source_index(table)
```

```python
def generate_random_color():
    
    """Случайная генерация цветов rgba

    Args:
        
    Returns:
        str: Строка со сгенерированными параметрами цвета
    """
    
    # сгенерим значение для каждого канала
    r, g, b = np.random.randint(255, size=3)
    return f'rgba({r}, {g}, {b}, 1)'
```

```python
def colors_for_sources(mode):
    
    """Генерация цветов rgba

    Args:
        mode (str): сгенерировать случайные цвета, если 'random', а если 'custom' - 
                    использовать заранее подготовленные
    Returns:
        dict: словарь с цветами, соответствующими каждому индексу
    """
    # словарь, в который сложим цвета в соответствии с индексом
    colors_dict = {}
    
    if mode == 'random':
        # генерим случайные цвета
        for label in df_comp['source'].unique():
            r, g, b = np.random.randint(255, size=3)            
            colors_dict[label] = f'rgba({r}, {g}, {b}, 1)'
            
    elif mode == 'custom':
        # присваиваем ранее подготовленные цвета
        colors = requests.get('https://raw.githubusercontent.com/rusantsovsv/senkey_tutorial/main/json/colors_senkey.json').json()
        for no, label in enumerate(table['source'].unique()):
            colors_dict[label] = colors['custom_colors'][no]
            
    return colors_dict
  
  
# генерю цвета из своего списка
colors_dict = colors_for_sources(mode='custom')
```

```python
def percent_users(sources, targets, values):
    
    """
    Расчет уникальных id в процентах (для вывода в hover text каждого узла)
    
    Args:
        sources (list): список с индексами source.
        targets (list): список с индексами target.
        values (list): список с "объемами" потоков.
        
    Returns:
        list: список с "объемами" потоков в процентах
    """
    
    # объединим источники и метки и найдем пары
    zip_lists = list(zip(sources, targets, values))
    
    new_list = []
    
    # подготовим список словарь с общим объемом трафика в узлах
    unique_dict = {}
    
    # проходим по каждому узлу
    for source, target, value in zip_lists:
        if source not in unique_dict:
            # находим все источники и считаем общий трафик
            unique_dict[source] = 0
            for sr, tg, vl in zip_lists:
                if sr == source:
                    unique_dict[source] += vl
                    
    # считаем проценты
    for source, target, value in zip_lists:
        new_list.append(round(100 * value / unique_dict[source], 1))
    
    return new_list
```

```python
def lists_for_plot(source_indexes=source_indexes, colors=colors_dict, frac=10):
    
    """
    Создаем необходимые для отрисовки диаграммы переменные списков и возвращаем
    их в виде словаря
    
    Args:
        source_indexes (dict): словарь с именами и индексами source.
        colors (dict): словарь с цветами source.
        frac (int): ограничение на минимальный "объем" между узлами.
        
    Returns:
        dict: словарь со списками, необходимыми для диаграммы.
    """
    
    sources = []
    targets = []
    values = []
    labels = []
    link_color = []
    link_text = []

    # проходим по каждому шагу
    for step in tqdm(sorted(table['step'].unique()), desc='Шаг'):
        if step + 1 not in source_indexes:
            continue

        # получаем индекс источника
        temp_dict_source = source_indexes[step]['sources_dict']

        # получаем индексы цели
        temp_dict_target = source_indexes[step+1]['sources_dict']

        # проходим по каждой возможной паре, считаем количество таких пар
        for source, index_source in tqdm(temp_dict_source.items()):
            for target, index_target in temp_dict_target.items():
                # делаем срез данных и считаем количество id            
                temp_df = table[(table['step'] == step)&(table['source'] == source)&(table['target'] == target)]
                value = len(temp_df)
                # проверяем минимальный объем потока и добавляем нужные данные
                if value > frac:
                    sources.append(index_source)
                    targets.append(index_target)
                    values.append(value)
                    # делаем поток прозрачным для лучшего отображения
                    link_color.append(colors[source].replace(', 1)', ', 0.2)'))
                    
    labels = []
    colors_labels = []
    for key in source_indexes:
        for name in source_indexes[key]['sources']:
            labels.append(name)
            colors_labels.append(colors[name])
            
    # посчитаем проценты всех потоков
    perc_values = percent_users(sources, targets, values)
    
    # добавим значения процентов для howertext
    link_text = []
    for perc in perc_values:
        link_text.append(f"{perc}%")
    
    # возвратим словарь с вложенными списками
    return {'sources': sources, 
            'targets': targets, 
            'values': values, 
            'labels': labels, 
            'colors_labels': colors_labels, 
            'link_color': link_color, 
            'link_text': link_text}
  

# создаем словарь
data_for_plot = lists_for_plot()
```

```python
def plot_senkey_diagram(data_dict=data_for_plot):    
    
    """
    Функция для генерации объекта диаграммы Сенкей 
    
    Args:
        data_dict (dict): словарь со списками данных для построения.
        
    Returns:
        plotly.graph_objs._figure.Figure: объект изображения.
    """
    
    fig = go.Figure(data=[go.Sankey(
        domain = dict(
          x =  [0,1],
          y =  [0,1]
        ),
        orientation = "h",
        valueformat = ".0f",
        node = dict(
          pad = 50,
          thickness = 15,
          line = dict(color = "black", width = 0.1),
          label = data_dict['labels'],
          color = data_dict['colors_labels']
        ),
        link = dict(
          source = data_dict['sources'],
          target = data_dict['targets'],
          value = data_dict['values'],
          label = data_dict['link_text'],
          color = data_dict['link_color']
      ))])
    fig.update_layout(title_text="Диаграмма Сэнкей", font_size=10)
    
    # возвращаем объект диаграммы
    return fig
  

# сохраняем диаграмму в переменную
senkey_diagram = plot_senkey_diagram()
```

### Строим воронки по основным сценариям в разрезе уникальных пользователей

Напишем функцию которая будет принемать данные и сторить воронку

```python
def voronka(df ,source1, source2 ,target):
    step1 = df[df['source'] == source1]['user_id'].nunique()
    step2 = df[(df['source'] == source1) & (df['target'] == source2)]['user_id'].nunique()
    step3 = df[(df['source'] == source2) & (df['target'] == target)]['user_id'].nunique()
    
    fig = go.Figure(go.Funnel(
        y=[source1, source2, target],  # Указываем ось y как названия шагов
        x=[step1, step2, step3],  # Указываем ось x как значения для каждого шага
        textinfo="value+percent previous"
    ))

    # Настройка внешнего вида воронки
    fig.update_layout(
        title="Воронка сценария",
        xaxis_title="Количество",
        yaxis_title="Шаги",
        showlegend=False
    )
    return fig.show()
```

## Проверка гипотез 

### Одни пользователи совершают действия tips_show и tips_click, другие – только tips_show. Проверить гипотезу, что конверсия в просмотр контактов различается у этих двух групп.

H0 - Коневерсии друх групп не различаются
<br>
H1 - Коневерсии различаються

```python
# Количество уникальных пользователей, совершивших tips_show и tips_click за одну сессию
tips_show_and_click_users = df[df['event_name'].isin(['tips_show' and 'tips_click'])]['user_id'].unique()

# Количество уникальных пользователей, совершивших только tips_show за одну сессию
tips_show_only_users = df[df['event_name'] == 'tips_show']['user_id'].unique()

# Количество уникальных пользователей, совершивших просмотр контактов
contacts_show_users_tips_show_and_click = df[(df['user_id'].isin(tips_show_and_click_users)) & (df['event_name'] == 'contacts_show')]['user_id'].nunique()
contacts_show_users_tips_show_only = df[(df['user_id'].isin(tips_show_only_users)) & (df['event_name'] == 'contacts_show')]['user_id'].nunique()

# Количество всех пользователей в каждой группе
total_users_tips_show_and_click = len(tips_show_and_click_users)
total_users_tips_show_only = len(tips_show_only_users)

# Проведение Z-теста для разности долей
z_stat, p_value = proportions_ztest([contacts_show_users_tips_show_and_click, contacts_show_users_tips_show_only],
                                     [total_users_tips_show_and_click, total_users_tips_show_only])

# Уровень значимости
alpha = 0.05
print('Всего пользователей которые совершили tips_show и tips_click:',total_users_tips_show_and_click)
print('Всего пользователей из группы tips_show и tips_click которые совершили ЦС:',contacts_show_users_tips_show_and_click)
print('Конверсия в ЦС',round((contacts_show_users_tips_show_and_click/total_users_tips_show_and_click) * 100, 2),'%')
print(' ')
print('Всего пользователей которые совершили tips_show:',total_users_tips_show_only)
print('Всего пользователей из группы tips_show которые совершили ЦС:',contacts_show_users_tips_show_only)
print('Конверсия в ЦС',round((contacts_show_users_tips_show_only/total_users_tips_show_only) * 100, 2),'%')
print(' ')
# Вывод результатов теста
print('P-значение: {:.20f}'.format(p_value))

# Проверка статистической значимости и вывод результата
if p_value < alpha:
    print('Отвергаем нулевую гипотезу: Конверсии в просмотр контактов различаются.')
else:
    print('Принимаем нулевую гипотезу: Конверсии в просмотр контактов не различаются.')
```

###  Различается ли конверсия в целевое действие пользователей, которые добавляют объявление в избраное, от тех кто не добавляет.

H0 - Конверсия в целеовое действие между группами не различается
<br>
H1 - Коневерсия различается

```python
# Количество уникальных пользователей, совершивших tips_show и tips_click за одну сессию
favorite_users = df[df['event_name'].isin(['favorites_add'])]['user_id'].unique()

# Количество уникальных пользователей, совершивших только tips_show за одну сессию
no_favorite_users = df[df['event_name'] != 'favorites_add']['user_id'].unique()

# Количество уникальных пользователей, совершивших просмотр контактов
contacts_show_favotire = df[(df['user_id'].isin(favorite_users)) & (df['event_name'] == 'contacts_show')]['user_id'].nunique()
contacts_show_no_favorite = df[(df['user_id'].isin(no_favorite_users)) & (df['event_name'] == 'contacts_show')]['user_id'].nunique()

# Количество всех пользователей в каждой группе
total_favorite_users = len(favorite_users)
total_no_favorite_users = len(no_favorite_users)

# Проведение Z-теста для разности долей
z_stat, p_value = proportions_ztest([contacts_show_favotire, contacts_show_no_favorite],
                                     [total_favorite_users, total_no_favorite_users])

# Уровень значимости
alpha = 0.05
print('Всего пользователей которые совершили favorites_add:',total_favorite_users)
print('Всего пользователей из группы favorites_add которые совершили ЦС:',contacts_show_favotire)
print('Конверсия в ЦС',round((contacts_show_favotire/total_favorite_users) * 100, 2),'%')
print(' ')
print('Всего пользователей которые не совершили favorites_add:',total_no_favorite_users)
print('Всего пользователей которые на добавляли в избраное и которые совершили ЦС:',contacts_show_no_favorite)
print('Конверсия в ЦС',round((contacts_show_no_favorite/total_no_favorite_users) * 100, 2),'%')
print(' ')
# Вывод результатов теста
print('P-значение: {:.20f}'.format(p_value))

# Проверка статистической значимости и вывод результата
if p_value < alpha:
    print('Отвергаем нулевую гипотезу: Конверсии в просмотр контактов различаются.')
else:
    print('Принимаем нулевую гипотезу: Конверсии в просмотр контактов не различаются.')
```
